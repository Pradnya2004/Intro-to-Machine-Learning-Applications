```python
import pandas as pd
df=pd.read_csv('train_weekthree.csv')
df
```

```python
#1 Training using Neural network
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, mean_squared_error
import tensorflow as tf
from sklearn.preprocessing import StandardScaler

# Data for Neural Network where I am choosing 'Number of times pregnant' as the target variable and others as features as it gives the lowest MSE.
X_nn = df.drop('Number of times pregnant', axis=1) #nn stands for neural network
y_nn = df['Number of times pregnant']

# Splitting the data into training and testing sets
X_train_nn, X_test_nn, y_train_nn, y_test_nn = train_test_split(X_nn, y_nn, test_size=0.2, random_state=42)

# Standardizing the features
scaler_nn = StandardScaler()
X_train_nn = scaler_nn.fit_transform(X_train_nn)
X_test_nn = scaler_nn.transform(X_test_nn)

# Training the neural network using tenserflow
model_nn = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(X_train_nn.shape[1],)),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])#Taking activation function as relu and sigmoid to set the input range b/w 0 and 1.

model_nn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model_nn.fit(X_train_nn, y_train_nn, epochs=10, batch_size=32, validation_data=(X_test_nn, y_test_nn))

# Predictions
y_pred_nn = model_nn.predict(X_test_nn)
y_pred_nn = (y_pred_nn > 0.5).astype(int).flatten()  # Converting probabilities to binary values

# Accuracy and mse of the neural network
accuracy_nn = accuracy_score(y_test_nn, y_pred_nn)
mse_nn = mean_squared_error(y_test_nn, y_pred_nn)
print('Neural Network Test Accuracy: ',accuracy_nn)
print('Neural Network Test MSE: ',mse_nn)
```

```python
#2
#Training using Decision Tree
from sklearn.tree import DecisionTreeClassifier

#Data for Decision tree and choosing 'Number of times pregnant' as the target variable.
X_dt = df.drop('Number of times pregnant', axis=1) #dt stands for decision tree
y_dt = df['Number of times pregnant']

# Splitting the data into training and testing sets
X_train_dt, X_test_dt, y_train_dt, y_test_dt = train_test_split(X_dt, y_dt, test_size=0.2, random_state=42)

# Creating a decision tree classifier
clf = DecisionTreeClassifier(random_state=42)

# Training the decision tree on the training data
clf.fit(X_train_dt, y_train_dt)

# Predictions
y_pred_dt = clf.predict(X_test_dt)

# Accuracy and MSE of the decision tree
accuracy_dt = accuracy_score(y_test_dt, y_pred_dt)
mse_dt = mean_squared_error(y_test_dt, y_pred_dt)
print('Decision Tree Test Accuracy:', accuracy_dt)
print('Decision Tree Test MSE: ',mse_dt)
```

```python
#3
# Ensemble predictions by averaging
y_pred_ensemble = np.round((y_pred_nn + y_pred_dt) / 2)

# Accuracy and MSE of the ensemble model
accuracy_ensemble = accuracy_score(y_test_dt, y_pred_ensemble)
mse_ensemble = mean_squared_error(y_test_dt, y_pred_ensemble)
print('Ensemble Test Accuracy: ',accuracy_ensemble)
print('Ensemble Test MSE: ',mse_ensemble)
```

```python
#5
#importing all the libraries again
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, mean_squared_error
import tensorflow as tf
from sklearn.preprocessing import StandardScaler

# Using sampling with replacement basically bootstrapping to increase the data size by 15%
original_size = len(df)
new_size = int(original_size * 1.15)
sampled_indices = np.random.choice(original_size, new_size, replace=True) #random allottment of data

# Creating a new df with the sampled data
df_sampled = df.iloc[sampled_indices]

# New data for neural network
X_nn = df_sampled.drop('Number of times pregnant', axis=1)
y_nn = df_sampled['Number of times pregnant']

# Splitting the data into training and testing sets
X_train_nn, X_test_nn, y_train_nn, y_test_nn = train_test_split(X_nn, y_nn, test_size=0.2, random_state=42)

# Standardizing the features
scaler_nn = StandardScaler()
X_train_nn = scaler_nn.fit_transform(X_train_nn)
X_test_nn = scaler_nn.transform(X_test_nn)

# Training the neural network
model_nn = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(X_train_nn.shape[1],)),#Taking activation function as relu and sigmoid to set the input range b/w 0 and 1.
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model_nn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model_nn.fit(X_train_nn, y_train_nn, epochs=10, batch_size=32, validation_data=(X_test_nn, y_test_nn))

# Predictions
y_pred_nn = model_nn.predict(X_test_nn)
y_pred_nn = (y_pred_nn > 0.5).astype(int).flatten()  # Converting probabilities to binary values

# Accuracy and MSE
accuracy_nn = accuracy_score(y_test_nn, y_pred_nn)
mse_nn = mean_squared_error(y_test_nn, y_pred_nn)
print('Neural Network Test Accuracy: ',accuracy_nn)
print('Neural Network Test MSE: ',mse_nn)

#New data for Decision Tree
X_dt = df_sampled.drop('Number of times pregnant', axis=1)
y_dt = df_sampled['Number of times pregnant']

# Splitting the data into training and testing sets
X_train_dt, X_test_dt, y_train_dt, y_test_dt = train_test_split(X_dt, y_dt, test_size=0.2, random_state=42)

# Creating a decision tree classifier
clf = DecisionTreeClassifier(random_state=42)

# Training the decision tree on the training data
clf.fit(X_train_dt, y_train_dt)

# Predictions
y_pred_dt = clf.predict(X_test_dt)

# Accuracy and MSE
accuracy_dt = accuracy_score(y_test_dt, y_pred_dt)
mse_dt = mean_squared_error(y_test_dt, y_pred_dt)
print('\nDecision Tree Test Accuracy:',accuracy_dt)
print('Decision Tree Test MSE:',mse_dt)

#Ensemble predictions by averaging
y_pred_ensemble = np.round((y_pred_nn + y_pred_dt) / 2)

# Accuracy and MSE of the ensemble model
accuracy_ensemble = accuracy_score(y_test_nn, y_pred_ensemble)
mse_ensemble = mean_squared_error(y_test_nn, y_pred_ensemble)
print('\nEnsemble Test Accuracy: ',accuracy_ensemble)
print('Ensemble Test MSE: ',mse_ensemble)
```
